### Step 7 - Performing text cleaning pipeline

```python
def text_cleaning(t):
    t_pun_rem = [ char for char in t if char not in string.punctuation]
    t_pun_rem = ''.join(t_pun_rem)
    t_clean = [word for word in t_pun_rem.split() if word.lower() not in stopwords.words('english')]
    return t_clean
    
```

```python
reviews_df_clean = reviews_df['verified_reviews'].apply(text_cleaning)
reviews_df_clean
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(analyzer = text_cleaning)
reviews_cv = cv.fit_transform(reviews_df['verified_reviews'])
reviews_cv.toarray()
```

```python
print(cv.get_feature_names())
```

```python
print(reviews_cv.toarray()[0])
```

```python
reviews_cv.shape
```

```python
reviews_df.drop(['verified_reviews'], axis=1, inplace=True)
reviews = pd.DataFrame(reviews_cv.toarray())
```

```python
reviews_df = pd.concat([reviews_df, reviews], axis=1)
```

```python
reviews_df
```

### Step 8 - Using Naive Bayes Classifier

```python
X = reviews_df.drop(['feedback'],axis=1)
```

```python
y = reviews_df['feedback']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

### Step 9 - Evaluating the NB classifier

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
y_predict_train = NB_classifier.predict(X_train)
y_predict_train
```

```python
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot = True)
```

```python
y_predict_test= NB_classifier.predict(X_test)
y_predict_test
```

```python
cm2 = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm2, annot = True)
```

We can see that the dataset is an imbalanced dataset because there are a of values which come under class 1.

```python
print(classification_report(y_test, y_predict_test))
```
